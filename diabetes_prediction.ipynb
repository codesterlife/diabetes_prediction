```python
import numpy as np
import pandas as pd

df = pd.read_csv("/home/codester/Desktop/diabetes.csv")
df
```

```python
df.describe(include="all")
```

```python
x = df.iloc[:, :-1] # input data
y = df.iloc[:, -1] # output data

y
```

```python
# converting the dataset into training and testing data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

```python
# scaling the dataset

from sklearn.preprocessing import StandardScaler

scaler  = StandardScaler()

scaler.fit(x_train)

scaler.transform(x_train)

scaler.transform(x_test)
```

```python
# training the KNN algorithm using the created datasets

from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors = 5)

classifier.fit(x_train, y_train)

y_prediction = classifier.predict(x_test)

y_prediction
```

```python
# checking the accuracy of currently trained model

from sklearn.metrics import accuracy_score

score = accuracy_score(y_test, y_prediction)

score * 100
```

```python
# Displaying the confusion matrix for the accuracy of the algorithm

from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

result = confusion_matrix(y_test, y_prediction)

label = [1, 0]

cmd = ConfusionMatrixDisplay(result, display_labels = label)

cmd.plot()
```
